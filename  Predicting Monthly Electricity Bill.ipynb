```python
#import libraries
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[3], [4], [6], [7], [8], [10], [12]])
y = np.array([50, 60, 80, 95, 110, 130, 150]) 
```

```python
# Initialize the linear regression model
model = LinearRegression()
```

```python
# Train the model on the dataset
model.fit(X, y) 
```

```python
# Now, let's predict the monthly bill for a household that uses electricity for 5 hours a day
predicted_bill_5_hours = model.predict(np.array([[5]]))
```

```python
# predict()applies the model to a new input
print("Predicted monthly bill for 5 hours of daily usage:",
predicted_bill_5_hours[0]) 
```

```python
slope = model.coef_[0] 
```

```python
# model.coef_ gives the slope of the line, showing bill increase per hour
intercept = model.intercept_
```

```python
# model.intercept_ is the y-intercept,showing the starting value of the bill
print("Slope (increase per additional hour):", slope)
print("Intercept (base bill):", intercept)
```

```python
import matplotlib.pyplot as plt
```

```python
#Plotting the data points
plt.scatter(X, y, color='blue', label='Data Points') 

# Creating predictions for the entire range of X values for plotting the regression line
X_range = np.array([[x] for x in range(2, 14)])
y_pred = model.predict(X_range) 
# Plotting the regression line
plt.plot(X_range, y_pred, color='red', label='Regression Line')

# Highlight the predicted value for 5 hours of usage
plt.scatter(5, predicted_bill_5_hours, color='green',label='Prediction for 5 hours', s=100, edgecolor='black')

# Adding labels and title
plt.title('Monthly Electricity Bill Prediction')
plt.xlabel('Average Daily Usage (hours)')
plt.ylabel('Monthly Bill (dollars)')
plt.legend()  
plt.grid(True)
# Show the plot
plt.show()
```
